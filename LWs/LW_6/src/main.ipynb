```python
import pandas
from os import getcwd


df = pandas.read_csv(getcwd() + "/LWs/LW_6/data/close_prices.csv")
# Акций 30 компаний на закрытии торгов за каждый день периода
X = df.loc[:, 'AXP':]
```

```python
for label in df:
    print(label)
```

```python
df.head(5)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(X.values)
```

```python
sum = 0
counter = 0
for value in pca.explained_variance_ratio_:
    counter += 1
    sum += value
    print("Было добавлено значение:", value, sep="\t")
    print("Текущая дисперсия:", sum, sep="\t", end="\n\n\n")
    if sum >= 0.9:
        break
print("Нужно компонент чтобы объяснить 90% дисперсии:", counter, sep="\t")
```

```python
newDF = pandas.DataFrame(pca.transform(X))
comp0 = newDF[0]
newDF.head(5)
```

```python
print(comp0)
```

```python
from numpy import corrcoef
df2 = pandas.read_csv(getcwd() + "/LWs/LW_6/data/djia_index.csv")
djia = df2['^DJI']
corr = corrcoef(comp0, djia)
print(djia)
```

```python
print(corr)
```

```python
_comp0 = pandas.Series(pca.components_[0])
print(_comp0, end="\n\n\n")
indexOfHigherValue = _comp0.sort_values(ascending=False).head(1).index[0]
company = X.columns[indexOfHigherValue]
print("Компания имеет наибольший вес в первой компоненте:", company, "индекс:", indexOfHigherValue, "вес", _comp0[indexOfHigherValue], sep="\t")
```
